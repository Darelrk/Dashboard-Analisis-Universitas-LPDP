```python
import pandas as pd
import io
import re
```

```python
nama_file_asli = 'the2025.csv'
nama_file_baru = 'the2025_cleaned.csv'

df = pd.read_csv(nama_file_asli)

kolom_yang_disimpan = ['Rank', 'Name Country/Region']
df_cleaned = df[kolom_yang_disimpan]

df_cleaned.to_csv(nama_file_baru, index=False)

print(df_cleaned.head())
```

```python
nama_file_asli = 'cwur2025.csv'
nama_file_hasil_csv = 'cwur2025_cleaned.csv'

df = pd.read_csv(nama_file_asli)

df_cleaned = df[['World Rank', 'Institution']].copy()

def clean_rank(rank_text):
    match = re.match(r'^\d+', str(rank_text))
    return match.group(0) if match else None

df_cleaned['World Rank'] = df_cleaned['World Rank'].apply(clean_rank)
df_cleaned.dropna(inplace=True)
df_cleaned['World Rank'] = df_cleaned['World Rank'].astype(int)

df_cleaned.to_csv(nama_file_hasil_csv, index=False)

print(df_cleaned.head())
```

```python
nama_file_asli = '2026QS.xlsx'
nama_file_baru = 'QS_2026_cleaned.csv'

df = pd.read_excel(nama_file_asli, skiprows=1)
df.columns = df.iloc[0]
df = df.drop(df.index[0]).reset_index(drop=True)

df_cleaned = df[['Rank', 'Name']]
df_cleaned = df_cleaned.rename(columns={'Rank': 'Rank', 'Name': 'Institution'})

df_cleaned.to_csv(nama_file_baru, index=False)

print(df_cleaned.head())
```
